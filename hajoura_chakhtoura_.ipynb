# Histo
**What It Is:**

A histogram shows the distribution of a numerical variable by dividing the data into bins (ranges) and counting how many data points fall into each bin.

**Why Use It:**

* To see the spread of data (e.g., age distribution).
* To identify skewness (whether data is symmetric or lopsided).

```python

fig, axs = plt.subplots(5, 3, figsize=(15, 15)) #namlo grid , 5 lignes , 3 C , e size total is 15 * 15 ( pour la visibilité )
axs = axs.ravel() #reshape to 1D
colors = sns.color_palette("husl", len(df.columns)) #generate N color using HUSL system

for i, (feature, color) in enumerate(zip(df.columns, colors)):#feature = current column name

    axs[i].hist(df[feature], color=color, edgecolor='black')#create a histo , for the curr feat
    axs[i].set_title(feature, fontsize=12)
    axs[i].set_xlabel('Value', fontsize=10)
    axs[i].set_ylabel('Frequency', fontsize=10)

for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])#remove un-used plts

plt.tight_layout()#adjusting the space between the plts
plt.show()
```

```python
# **Cell 7: Visualizations - Comparative Histograms Based on Target Variable**
# Comparing distributions of 'age', 'chol', and 'trtbps' for different target classes.

def plot_comparative_histograms(df, feature, target, color1='green', color2='red'):
    """
    Plots comparative histograms for a given feature based on the target variable.

    Parameters:
    - df: pandas DataFrame containing the data
    - feature: Feature to plot
    - target: Target variable for comparison
    - color1: Color for target = 0
    - color2: Color for target = 1
    """
    plt.figure(figsize=(10, 6))
    sns.histplot(df[df[target] == 0][feature], color=color1, kde=True, stat="density", bins=30, label='No Heart Attack')
    sns.histplot(df[df[target] == 1][feature], color=color2, kde=True, stat="density", bins=30, label='Heart Attack')
    plt.title(f'{feature.capitalize()} Distribution by Heart Attack Status', fontsize=15)
    plt.xlabel(feature.capitalize(), fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.legend()
    plt.show()

# Plotting comparative histograms for 'age', 'chol', and 'trtbps'
comparative_features = ['age', 'chol', 'trtbps']
for feature in comparative_features:
    plot_comparative_histograms(df, feature, 'output')
```

# Data Preprocessing


```python
from sklearn.model_selection import train_test_split
def dataset(X,y):
    train_full_X, val_X, train_full_y, val_y = train_test_split(X, y,test_size=0.2,random_state = 0)

# Use the same function above for the validation set
    train_X, test_X, train_y, test_y = train_test_split(train_full_X, train_full_y, test_size=0.25,random_state = 0)
    return (train_X, val_X, train_y, val_y)
```

Minmax scaling

```python
from sklearn.preprocessing import MinMaxScaler

def scale(X):
    scaler = MinMaxScaler()
    scaler.fit(X)
    return X
```

**decision tree here**

with all features

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
model=[]

X= df.drop('output',axis=1)
y= df.output

X= scale(X)
X_train,X_val,y_train,y_val= dataset(X,y)

tree_model= DecisionTreeClassifier(random_state=1)
tree_model.fit(X_train, y_train)
preds= tree_model.predict(X_val)

a='Decision Tree with all features'
acc=accuracy_score(y_val,preds)*100
roc=roc_auc_score(y_val,preds)

print('ACCURACY: ',acc)
print('AUC score: ',roc)

model.append((a,round(acc,4),round(roc,4)))
```

with some parameter tuning

```python
features=['thall','caa','cp','oldpeak','exng','chol','thalachh']
X= df[features]
y= df.output

depth = [1, 2, 3, 4, 5, 6, 10, 15, 20, None]
leaf=[1, 5, 10, 15, 20, 500, 100, 200]
scores=[]
for d in depth:
      for l in leaf:
        dmodel = DecisionTreeClassifier(random_state=0,max_depth=d,min_samples_leaf=l)
        dmodel.fit(X_train, y_train)

        pred1 = dmodel.predict(X_val)
        acc= accuracy_score(y_val,pred1)*100
        auc = roc_auc_score(y_val,pred1)
        #print('%4s %s -> auc %.3f accuracy  %.2f' % (d, l, auc, acc))

        scores.append((d,l,auc))
```

```python
columns = ['max_depth', 'min_samples_leaf', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)

pivot2 = df_scores.pivot(index='min_samples_leaf', columns=['max_depth'], values=['auc'])
pivot2.round(3)
```

neural network part :

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, roc_auc_score

# Define the neural network architecture
nn_model = Sequential()

# Input layer (input shape is based on the number of features in X_train)
nn_model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))

# Hidden layers
nn_model.add(Dense(32, activation='relu'))
nn_model.add(Dense(16, activation='relu'))

# Output layer (for binary classification, so 1 unit with sigmoid activation)
nn_model.add(Dense(1, activation='sigmoid'))

# Compile the model
nn_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = nn_model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_val, y_val), verbose=1)

# Make predictions on the validation set
predictions_nn = nn_model.predict(X_val)
predictions_nn = (predictions_nn > 0.5).astype(int).flatten()

# Evaluate the model
acc_nn = accuracy_score(y_val, predictions_nn) * 100
roc_nn = roc_auc_score(y_val, predictions_nn)

print(f'Neural Network ACCURACY: {acc_nn}')
print(f'Neural Network AUC score: {roc_nn}')

# Append results to the model comparison DataFrame
a = 'Neural Network'
new_row = pd.DataFrame([[a, round(acc_nn, 4), round(roc_nn, 4)]], columns=['Model', 'Accuracy', 'AUC'])
model = pd.concat([model, new_row], ignore_index=True)

# Optionally, plot the training history for accuracy and loss
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
